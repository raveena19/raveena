```python
Common techniques used in univariate analysis include:
1.Descriptive Statistics
2.Frequency Distributions
3.Box Plots
4.Kernel Density Estimation (KDE)
5.Probability Plots
```

```python
#1.descriptive statistics
import pandas as pd
import seaborn as sns

# Create DataFrame
data = {'points': [1, 1, 2, 3.5, 4, 4, 4, 5, 5, 6.5, 7, 7.4, 8, 13, 14.2],
        'assists': [5, 7, 7, 9, 12, 9, 9, 4, 6, 8, 8, 9, 3, 2, 6],
        'rebounds': [11, 8, 10, 6, 6, 5, 9, 12, 6, 6, 7, 8, 7, 9, 15]}
df = pd.DataFrame(data)

```

```python
# Descriptive statistics
summary_statistics = df.describe()
print("Summary Statistics:\n", summary_statistics)

```

```python
# Median
median_values = df.median()
print("\nMedian Values:\n", median_values)


```

```python
# Mode
mode_values = df.mode().iloc[0]
print("\nMode Values:\n", mode_values)

```

```python
# Variance
variance_values = df.var()
print("\nVariance Values:\n", variance_values)

```

```python
# Standard Deviation
std_dev_values = df.std()
print("\nStandard Deviation Values:\n", std_dev_values)
```

```python
#frequency distributions
import matplotlib.pyplot as plt
# Frequency distribution for 'points'
plt.figure(figsize=(8, 6))
plt.hist(df['points'], bins=10, edgecolor='black')
plt.title('Frequency Distribution of Points')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()


```

```python
# Frequency distribution for 'assists'
plt.figure(figsize=(8, 6))
plt.hist(df['assists'], bins=10, edgecolor='black')
plt.title('Frequency Distribution of Assists')
plt.xlabel('Assists')
plt.ylabel('Frequency')
plt.show()

```

```python
# Frequency distribution for 'rebounds'
plt.figure(figsize=(8, 6))
plt.hist(df['rebounds'], bins=10, edgecolor='black')
plt.title('Frequency Distribution of Rebounds')
plt.xlabel('Rebounds')
plt.ylabel('Frequency')
plt.show()
```

```python
#box plot
# Box plot for 'points'
plt.figure(figsize=(8, 6))
sns.boxplot(x='points', data=df)
plt.title('Box Plot of Points')
plt.show()

```

```python
# Box plot for 'assists'
plt.figure(figsize=(8, 6))
sns.boxplot(x='assists', data=df)
plt.title('Box Plot of Assists')
plt.show()


```

```python
# Box plot for 'rebounds'
plt.figure(figsize=(8, 6))
sns.boxplot(x='rebounds', data=df)
plt.title('Box Plot of Rebounds')
plt.show()
```

```python
#kernal density estimation

# KDE plot for 'points'
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df['points'], fill=True)
plt.title('Kernel Density Estimation (KDE) of Points')
plt.xlabel('Points')
plt.ylabel('Density')
plt.show()

```

```python
# KDE plot for 'rebounds'
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df['rebounds'], fill=True)
plt.title('Kernel Density Estimation (KDE) of Rebounds')
plt.xlabel('Rebounds')
plt.ylabel('Density')
plt.show()
```

```python
# KDE plot for 'assists'
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df['assists'], fill=True)
plt.title('Kernel Density Estimation (KDE) of Assists')
plt.xlabel('Assists')
plt.ylabel('Density')
plt.show()


```

```python
#probability plots
from scipy.stats import probplot
# Probability plot for 'points'
plt.figure(figsize=(8, 6))
probplot(df['points'], plot=plt)
plt.title('Probability Plot of Points')
plt.show()



```

```python
# Probability plot for 'assists'
plt.figure(figsize=(8, 6))
probplot(df['assists'], plot=plt)
plt.title('Probability Plot of Assists')
plt.show()

```

```python
4. Write a NumPy program to convert a Python dictionary to a NumPy ndarray. Sample Output:
Original dictionary:
{'column0': {'a': 1, 'b': 0.0, 'c': 0.0, 'd': 2.0},


'column1': {'a': 3.0, 'b': 1, 'c': 0.0, 'd': -1.0},
'column2': {'a': 4, 'b': 1, 'c': 5.0, 'd': -1.0},
'column3': {'a': 3.0, 'b': -1.0, 'c': -1.0, 'd': -1.0}}
Type: <class 'dict'>
ndarray:
[[ 1. 0. 0. 2.]
[ 3. 1. 0. -1.]
[ 4. 1. 5. -1.]
[ 3. -1. -1. -1.]]
Type: <class 'numpy.ndarray'>
```

```python
import numpy as np

# Original dictionary
original_dict = {'column0': {'a': 1, 'b': 0.0, 'c': 0.0, 'd': 2.0},
                 'column1': {'a': 3.0, 'b': 1, 'c': 0.0, 'd': -1.0},
                 'column2': {'a': 4, 'b': 1, 'c': 5.0, 'd': -1.0},
                 'column3': {'a': 3.0, 'b': -1.0, 'c': -1.0, 'd': -1.0}}

# Convert the dictionary to a NumPy ndarray
result_array = np.array([[v[subkey] for subkey in ['a', 'b', 'c', 'd']] for k, v in original_dict.items()])

# Display the result
print("Original dictionary:")
print(original_dict)
print("Type:", type(original_dict))

print("\nndarray:")
print(result_array)
print("Type:", type(result_array))
```

```python
5.Write a Pandas program to select the rows where the number of attempts in the examination is
greater than 2. Sample Python dictionary data and list labels:
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew',
'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
Expected Output:
Number of attempts in the examination is greater than 2:
name score attempts qualify
b Dima 9.0 3 no
d James NaN 3 no

f Michael 20.0 3 yes
```

```python
import pandas as pd
import numpy as np

# Sample data
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Create DataFrame
df = pd.DataFrame(exam_data, index=labels)

# Select rows where the number of attempts is greater than 2
result_df = df[df['attempts'] > 2]

# Display the result
print("Number of attempts in the examination is greater than 2:\n", result_df)
```
